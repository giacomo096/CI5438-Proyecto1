```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv('CarDekho.csv')
```

```python
df2 = df[["Make","Year","Kilometer","Fuel_Type","Transmission","Owner","Seating_Capacity","Fuel_Tank_Capacity","Price"]]
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2["Seating_Capacity"] = df2["Seating_Capacity"].fillna(5)
```

```python
df2["Fuel_Tank_Capacity"] = df2["Fuel_Tank_Capacity"].fillna(35)
```

```python
df2.describe()
```

```python
df2.info()
```

```python
df2["Owner"].value_counts()
```

```python
df2
```

```python
s = df2["Transmission"]=="Manual"
s = s.replace([True,False],[1,0])
```

```python
s2 = df2["Transmission"]=="Automatic"
s2 = s2.replace([True,False],[1,0])
```

```python
df2["Automatic"] = s2
df2["Manual"] = s
```

```python
df2 = df2.drop("Transmission", axis = 1)
```

```python
df2
```

```python
l = []
for x in df2["Make"].unique():
    y = df2.loc[df2["Make"] == x ]
    l.append((y["Price"].mean(),x))

print(l)
```

```python
l.sort()
i = 1
li = []
lj = []
index = []
for x,y in l:
    li.append(y)
    lj.append(i)
    index.append((y,i))
    i = i+1
```

```python
index
```

```python
df2["Make"] = df2["Make"].replace(li,lj)
```

```python
df2
```

```python
df2["Owner"] = df2["Owner"].replace(["Third","Fourth","4 or More", "UnRegistered Car"],"Others")
```

```python
s3 = df2["Owner"]=="First"
s3 = s3.replace([True,False],[1,0])
s4 = df2["Owner"]=="Second"
s4 = s4.replace([True,False],[1,0])
s5 = df2["Owner"]=="Others"
s5 = s5.replace([True,False],[1,0])
```

```python
df2["First_O"] = s3
df2["Second_O"] = s4
df2["Others_O"] = s5
df2 = df2.drop("Owner", axis = 1)
```

```python
laux = ['CNG', 'LPG', 'Electric', 'CNG + CNG',
       'Hybrid', 'Petrol + CNG', 'Petrol + LPG']
```

```python
df2["Fuel_Type"] = df2["Fuel_Type"].replace(laux,"Others")
```

```python
s6 = df2["Fuel_Type"]=="Petrol"
s6 = s6.replace([True,False],[1,0])
s7 = df2["Fuel_Type"]=="Diesel"
s7 = s7.replace([True,False],[1,0])
s8 = df2["Fuel_Type"]=="Others"
s8 = s8.replace([True,False],[1,0])
```

```python
s8.value_counts()
```

```python
df2["Petrol"] = s6
df2["Diesel"] = s7
df2["Others_F"] = s8
df2 = df2.drop("Fuel_Type", axis = 1)
```

```python
df2.to_csv("clean_data.csv")
```
